```python
import numpy as np
import matplotlib.pyplot as plt

# モンテカルロ法の試行回数
n_tot = 10000
# n_tot×2の配列の一様乱数を生成
x1 = np.random.rand(n_tot,2)

# 円の中に入った乱数の個数の初期値(ゼロ)
n_in = 0
plt.figure(figsize=(5,5))

# 半径1の円の内側に入る場合は緑のプロットで、円の外の場合は青で示す
for i in range(n_tot):
    if (x1[i,0]**2+x1[i,1]**2)**0.5 < 1:  #  半径1の円の内側に入る場合
        plt.plot(x1[i,0],x1[i,1],color='green', marker='x')
        n_in += 1
    else:                                 #  半径1の円の外側に入る場合
        plt.plot(x1[i,0],x1[i,1],color='blue', marker='x')
print('Estimate value of pi =',4*(n_in/n_tot))
```

```python
import numpy as np
import matplotlib.pyplot as plt

#  モンテカルロ法の試行回数のリスト(試行回数100から1000000回まで)
n_tot = [100,1000,10000,100000,1000000]

#  各n_totのケースの実行
for i in n_tot:
    # 乱数列の作成
    x1 = np.random.rand(i,2)
    # 全ての乱数列を2乗
    x2 = x1**2
    # 乱数列の2乗和の作成
    x3 = np.sum(x2,axis=-1)
    # 乱数列のルート
    x4 = x3**0.5
    # 2乗和のルートが1未満の数の個数の計算
    x5 = np.count_nonzero(x4 < 1)
    plt.plot(i,4*x5/i,color='green',marker='o')
x=[0,1000000]
y=[np.pi, np.pi]
plt.plot(x, y, color = 'red',linestyle='solid', label='true value')
plt.ylim(0,3.5);
plt.title('Comparision of try count and estimate value of pi')
plt.xlabel('try count')
plt.ylabel('Estimate value of pi')
plt.xscale('log')
plt.legend()
```
