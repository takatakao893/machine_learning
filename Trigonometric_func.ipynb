```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 3/4*np.cos(x/4)

def g(x):
    return (1+f(x))*np.cos(4*x)


# x
x = np.arange(-20, 20.1, 0.1)

plt.plot(x, f(x), color='blue')
plt.plot(x, g(x), color='red')
plt.legend(loc="upper left") # 
plt.ylim(-2,2) # y
plt.title(r'blue: $\frac{3}{4}\cos(\frac{x}{4})$     red: $(1+\frac{3}{4}\cos(\frac{x}{4}))\cos(4x)$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return -np.sin(x)/(2+np.cos(x))

def g(x):
    return np.sin(x)/(2+np.cos(x))

def h(x):
    return np.cos(x)/(2+np.sin(x))

# x
x = np.arange(-10, 10.1, 0.1)
plt.plot(x, f(x), color='green')
plt.plot(x, g(x), color='blue')
plt.plot(x, h(x), color='red')
plt.legend(loc="upper left")
plt.title(r'green: $\frac{-\sin(x)}{2+\cos(x)}$     blue: $\frac{\sin(x)}{2+\cos(x)}$       red: $\frac{\cos(x)}{2+\sin(x)}$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.ylim(-2.5,2.5)
plt.grid(True)
plt.show()
```
