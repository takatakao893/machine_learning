```python
import numpy as np
import matplotlib.pyplot as plt

# 2
def f(x, y):
    return x**2 + y**2

# xy
x = np.arange(-3, 4)
y = np.arange(-3, 4)

# 
X, Y = np.meshgrid(x,y)

# 2
Z = X**2 + Y**2

# (x, y, z)
px = 1.0
py = 2.0

# z
pz = f(px, py)

# (dx, dy)
dpx = 2 * px
dpy = 2 * py

# 
bx = f(px, py) - dpx * px
by = f(px, py) - dpy * py

# 
tangent_line_x = dpx * x + bx
tangent_line_y = dpy * y + by

# 
tangent_plane = dpx * (X-px) + dpy * (Y-py) + f(px,py)

# 
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(projection='3d')
ax.plot_wireframe(X, Y, Z, label='f(x,y)')
ax.scatter(px, py, pz, s=100, c='black', label='(x,y,z)') # 
ax.plot_wireframe(X, Y, tangent_plane, color='orange', alpha=0.5) # 
ax.plot(x, np.repeat(py, len(x)), tangent_line_x, 
        color='red',linewidth=2.5, label='$f_x(x, y)$') # x0
ax.plot(np.repeat(px, len(y)), y, tangent_line_y,
       color='green', linewidth=2.5, label='$f_y(x, y)$') # x1
ax.set_xlabel('x') # x
ax.set_ylabel('y') # y
ax.set_zlabel('z') # z
ax.legend()
```
